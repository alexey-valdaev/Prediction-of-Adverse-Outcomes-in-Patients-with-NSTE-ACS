```python
import numpy as np

def convert_to_numeric(value):
    try:
        return pd.to_numeric(value)
    except ValueError:
        # если значение содержит "<", мы берем число после этого символа
        if '<' in value:
            return pd.to_numeric(value[1:])
        else:
            return np.nan

#df['СРБ'] = df['СРБ'].apply(convert_to_numeric)
df['Тропонин_2'] = df['Тропонин_2'].apply(convert_to_numeric)
df['Тропонин_1'] = df['Тропонин_1'].apply(convert_to_numeric)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, RobustScaler
import numpy as np

X = df.drop(['Исход'], axis=1)
y = df['Исход']
```

```python
# Создадим объекты масштабирования и нормализации
min_max_scaler = MinMaxScaler()
robust_scaler = RobustScaler()

columns_MMS = []
columns_RS = ['Моно_абс', 'Hb', 'MPV', 'SII', 'Окклюзия или Ствол стеноз']

ct_O = ColumnTransformer(
    [("min_max_scaling", min_max_scaler, columns_MMS),
     ("robust_scaling", robust_scaler, columns_RS)],
    remainder="passthrough",
)

X_train_transformed = ct_O.fit_transform(X_train)
X_test_transformed = ct_O.transform(X_test)
```

## Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import make_scorer, f1_score, classification_report, confusion_matrix

# Define the models
logreg = LogisticRegression(class_weight='balanced', max_iter=5000)

# Define the hyperparameters for GridSearchCV
param_grid_logreg = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}

# Define the scoring metric
scorer = make_scorer(f1_score)

# Define the GridSearchCV objects
grid = GridSearchCV(estimator=logreg, param_grid=param_grid_logreg, scoring=scorer, cv=5)

# Fit the GridSearchCV objects to the data and predict the test labels
grid.fit(X_train_transformed, y_train)
y_pred = grid.predict(X_test_transformed)
print(f"Classification report for Logistic Regression:\n")
print(classification_report(y_test, y_pred))

# Вычисляем матрицу ошибок
cm = confusion_matrix(y_test, y_pred)

# Выводим матрицу ошибок
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.ylabel('Истинное значение')
plt.xlabel('Предсказанное значение')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc

# Предсказываем вероятности классов
y_pred_proba = grid.predict_proba(X_test_transformed)

# Вычисляем ROC AUC
roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
print(f"ROC AUC: {roc_auc}")

# Вычисляем PR AUC
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba[:, 1])
pr_auc = auc(recall, precision)
print(f"PR AUC: {pr_auc}")
```

```python
from sklearn.metrics import roc_curve, auc

# Строим ROC кривую
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:, 1])
roc_auc = auc(fpr, tpr)

# Визуализация ROC кривой
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

print("AUC: ", roc_auc)
```

```python
# Получение лучшей модели
best_model = grid.best_estimator_

# Получение важности признаков
importances = best_model.coef_[0]

# Создание DataFrame для отображения важности признаков
feature_importances = pd.DataFrame({'feature': X.columns, 'importance': importances})

# Сортировка DataFrame по важности признаков в порядке убывания
feature_importances = feature_importances.sort_values('importance', ascending=False)
```
